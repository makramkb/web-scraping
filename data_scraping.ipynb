```python
# importing dependencies

from bs4 import BeautifulSoup
from splinter import Browser
```

```python
# Use chrome 

browser = Browser('chrome')
```

```python
# Visit the website 

url = 'https://static.bc-edx.com/data/web/mars_news/index.html'
browser.visit(url)
```

```python
# Parse html

html = browser.html
soup = BeautifulSoup(html,'html.parser')
```

```python
# Test the data

soup.body
```

```python
# Get all articles from the page

soup_article = soup.find_all('div',class_='article_teaser_body')
for article in soup_article:
    print(article.text)
```

```python
# Get all dates

soup_date = soup.find_all('div',class_='list_date')
for date in soup_date:
    print(date.text)
```

```python
# Get all img links

soup_img = soup.find_all('img')
for image in soup_img:
    print(image['src'])
```

```python
# Find all divs with class col-md-3

all_divs = soup.find_all('div',class_='col-md-3')
all_divs
```

```python
# Loop through all divs to extract the h5 text

text_list = []
for i in range(len(all_divs)):
    all_h5 = all_divs[i].find_all('h5')
    # Skip the div that has no h5 and append the ones that exist to the text_list list
    if len(all_h5)>0:
        all_text = [text_list.append(h5.text) for h5 in all_h5]

text_list
```

```python
# Loop through all divs to extract the list text 

all_li_list = []
for i in range(len(all_divs)):
    all_li = all_divs[i].find_all('li')
    # Add the list only if list exists
    if len(all_li)>0:
        [all_li_list.append(li.text) for li in all_li]

print(all_li_list)
```
